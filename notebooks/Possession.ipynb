# Defining Possessions  

A **possession** starts when one team gains control of the basketball and ends when that team gives up control of the basketball.  

Examples of how teams give up control:  
- Made field goals or free throws in which other team inbounds ball  
- Defensive rebounds 
- Turnovers  

### Possessions are approximately the same for two teams in a single game (>2 for non OT games) therefore useful for evaluating efficiency of teams and players.
