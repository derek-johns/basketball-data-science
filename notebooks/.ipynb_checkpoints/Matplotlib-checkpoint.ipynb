```python
import matplotlib.pyplot as plt
```

```python
# Create a figure
fig = plt.figure()
# Assign axes
ax = plt.axes()
# Configure grid lines
plt.grid()
# Display
plt.show()
```

## Properties of keyword arguments for grid attribute.  
`b`: boolean (True/False), show grid lines
* If any other args present b=False by default  

`color`: assigns color to grid  
`alpha`: change opacity of grid (0 to disappear)  
`ls`: set style of grid lines
* '-', '--', '-.', ':', '' 

`lw`: set width of grid lines

```python
plt.grid(b=True, color='aqua', alpha=0.3, ls='-.', lw=2)
plt.show()
```

### Plot lines can be styled the same way.

```python
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), color='Indigo', ls='--', lw=3)
plt.grid(color='Aqua', ls='-.', lw=2, alpha=0.3)
plt.show()
```

## Axis Labels

```python
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=20)
plt.show()
```

## Tick Parameters

```python
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='b', colors='r', grid_color='gray', grid_alpha=0.5)
plt.grid()
plt.show()
```

## Change Fig Size

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), color='Indigo', ls='--', lw=3)
plt.grid(color='Aqua', ls='-.', lw=2, alpha=0.3)
plt.rcParams['figure.figsize'] = (10, 5)
plt.show()
```
